```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import trim_mean, kurtosis, norm
from scipy.stats.mstats import mode, gmean, hmean
```

```python
histo_bins_count = 20
sns.set()
```

```python
df_via = pd.read_csv('./via.csv')
df_nella = pd.read_csv('./nella.csv')


print('VIA')
print(df_via.describe())
print('===================================================================')
print('NELLA')
print(df_nella.describe())
```

```python
print('danceability_median', 'via', df_via['danceability'].median(), 'nella', df_nella['danceability'].median())
print('danceability_min', 'via', df_via['danceability'].min(), 'nella', df_nella['danceability'].min())
print('danceability_max', 'via', df_via['danceability'].max(), 'nella', df_nella['danceability'].max())
print('danceability_mean', 'via', df_via['danceability'].mean(), 'nella', df_nella['danceability'].mean())
print('danceability_std', 'via', df_via['danceability'].std(), 'nella', df_nella['danceability'].std())
print('danceability_skew', 'via', df_via['danceability'].skew(), 'nella', df_nella['danceability'].skew())
```

```python
df_via['danceability'].hist(bins=histo_bins_count)
plt.show()
```

```python
mu, sigma = 0.5, 0.1

# Create the bins and histogram
count, bins, ignored = plt.hist(df_via['danceability'], histo_bins_count, density=True)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ),       linewidth=3, color='y')
plt.xlabel('danceability of via')
plt.show()
```

```python
df_nella['danceability'].hist(bins=histo_bins_count)
plt.show()
```

```python
mu, sigma = 0.5, 0.1

# Create the bins and histogram
count, bins, ignored = plt.hist(df_nella['danceability'], histo_bins_count, density=True)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ),       linewidth=3, color='y')
plt.xlabel('danceability of nella')
plt.show()
```

```python
print('valence_min', 'via', df_via['valence'].min(), 'nella', df_nella['valence'].min())
```

```python
df_via['valence'].hist(bins=histo_bins_count)
plt.show()
```

```python
df_nella['valence'].hist(bins=histo_bins_count)
plt.show()
```

```python
_ = sns.swarmplot(x='danceability', y='valence', data=df_via)

# Label the axes
_ = plt.xlabel('danceability via')
_ = plt.ylabel('valence via')

# Show the plot
plt.show()
```

```python
_ = sns.swarmplot(x='danceability', y='valence', data=df_nella)

# Label the axes
_ = plt.xlabel('danceability nella')
_ = plt.ylabel('valence nella')

# Show the plot
plt.show()
```

```python
percentile = np.array([10,20,50,70,90])
print('percentile', percentile)
ptiles_via = np.percentile(df_via['danceability'], percentile)
x_via = np.sort(df_via['danceability'])
y_via = np.arange(1, len(x_via)+1) / len(x_via)
_ = plt.plot(x_via, y_via, marker = '.', linestyle='none')
_ = plt.plot(ptiles_via, percentile/100, marker='D', color='yellow', linestyle='none')

ptiles_nella = np.percentile(df_nella['danceability'], percentile)
x_nella = np.sort(df_nella['danceability'])
y_nella = np.arange(1, len(x_nella)+1) / len(x_nella)
_ = plt.plot(x_nella, y_nella, marker = '.', linestyle='none')
_ = plt.plot(ptiles_nella, percentile/100, marker='D', color='blue', linestyle='none')

plt.legend(['via', 'ptiles_via', 'nella', 'ptiles_nella'])
_ = plt.xlabel('danceability')
_ = plt.ylabel('ecdf')
plt.margins(0.02)
plt.show()
```

```python
x_via = np.sort(df_via['valence'])
y_via = np.arange(1, len(x_via)+1) / len(x_via)
_ = plt.plot(x_via, y_via, marker = '.', linestyle='none')

x_nella = np.sort(df_nella['valence'])
y_nella = np.arange(1, len(x_nella)+1) / len(x_nella)
_ = plt.plot(x_nella, y_nella, marker = '.', linestyle='none')

plt.legend(['via', 'nella'])
_ = plt.xlabel('valence')
_ = plt.ylabel('ecdf')
plt.margins(0.02)
plt.show()
```
